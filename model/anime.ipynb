```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
anime_data['rating'] = anime_data['rating'].fillna(anime_data['rating'].mode()[0])

anime_data['rating'].value_counts()
```

```python
anime_data.isnull().sum()
```

```python
anime_data.columns
```

```python
anime_data['synopsis_genres'] = anime_data['synopsis'] + ' ' + anime_data['genres']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
anime_data['synopsis_genres'].shape
```

```python
# Sample small selection of data

anime_data_sample = anime_data.sample(frac=0.6, random_state=42)

print(anime_data.shape)
print(anime_data_sample.shape)
```

```python
vectorizer = TfidfVectorizer()

X_transformed = vectorizer.fit_transform(anime_data_sample['synopsis_genres'])

print(X_transformed)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim = cosine_similarity(X_transformed, X_transformed)

cosine_sim_df = pd.DataFrame(cosine_sim, index=anime_data_sample['synopsis_genres'], columns=anime_data_sample['synopsis_genres'])

print(cosine_sim_df)
```

```python
np.save('cosine_sim.npy', cosine_sim)
```
